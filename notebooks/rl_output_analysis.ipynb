### Notebook to gather output of different models and analyse performance

#### Imports and Installations

```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

#### Access Output

```python
# define folder to gaher data from
input_dir = '../../sumo-rl/outputs/'
```

```python
sumo-rl/outputs/big-intersection
```

```python
for file in os.listdir(input_dir + 'big-intersection/'):
    main_df = pd.DataFrame()
    main_df = pd.read_csv(input_dir + 'big-intersection/' + file)
    main_df = main_df[['step', 'system_total_waiting_time']]
    plt.xlabel('Time step (seconds)')
    plt.ylabel('Total waiting time (s)')
    plt.plot(main_df['step'], main_df['system_total_waiting_time'])
```
